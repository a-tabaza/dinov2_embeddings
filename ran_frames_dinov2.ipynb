```python
import numpy as np
import json
from uuid import uuid4
import pandas as pd
from tqdm import tqdm
```

```python
embeddings = json.loads(open('dinov2_ran_embeddings.json').read())
```

```python
df = pd.DataFrame.from_dict(embeddings, orient='index')
```

```python
df['frame'] = df.index
```

```python
df['frame'] = df['frame'].apply(lambda x: x.split('/')[-1].split('.')[0])
```

```python
df.index = df['frame']
```

```python
df.drop('frame', axis=1, inplace=True)
```

```python
data = [key for key, _ in embeddings.items()]
data = [data.split('/')[-1] for data in data]
data = [data[:4] for data in data]
```

```python
import chromadb
client = chromadb.PersistentClient(path="db")
```

```python
collection = client.get_or_create_collection(name="ran_frames_dinov2")
```

```python
data
```

```python
for idx, data_ in tqdm(enumerate(data)):
    collection.add(
        documents=[f'frame_{data_}'],
        embeddings=[df.loc[data_].tolist()],
        metadatas=[{'frame': str(data_)}],
        ids=[data_]
    )
```
